```python
import pandas as pd

rerank_results_path = "reranking_retriever_evaluation_results.xlsx"
retriever_results_path = "retriever_evaluation_results.xlsx"

rerank_results = pd.read_excel(rerank_results_path)
retriever_results = pd.read_excel(retriever_results_path)

rerank_results.head(), retriever_results.head()
```

```python
rerank_results.round(2).to_csv("reranking_retriever_evaluation_results.csv", index=False)
retriever_results.round(2).to_csv("retriever_evaluation_results.csv", index=False)
```

```python
rerank_results_all = rerank_results[
    ["model", "dataset", "settings"] + [f"Accuracy@{i}" for i in range(1, 11)]
].rename(columns={f"Accuracy@{i}": f"Accuracy@{i}_After_Rerank" for i in range(1, 11)})

retriever_results_all = retriever_results[
    ["model", "dataset", "settings"] + [f"Accuracy@{i}" for i in range(1, 11)]
]

comparison_results_all = pd.merge(
    retriever_results_all, rerank_results_all, on=["model", "dataset", "settings"]
)

ordered_columns = ["model", "dataset", "settings"] + [
    col
    for pair in zip(
        [f"Accuracy@{i}" for i in range(1, 11)],
        [f"Accuracy@{i}_After_Rerank" for i in range(1, 11)],
    )
    for col in pair
]

comparison_results_all = comparison_results_all[ordered_columns]

comparison_results_all
```

```python
comparison_results_all.round(2).to_csv("compare_reranked_results.csv", index=False)
```
